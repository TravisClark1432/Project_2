```python
# Drop rows with missing values in the 'genres' column so we can find random samples
df_dropped_na = df.dropna(subset=['genres'])
```

```python
# Filter rows containing 'Action' in the 'genres' column
action = df_dropped_na[df_dropped_na['genres'].str.contains('Action')]

# Getting a random sample of 40 movies containing "action"
random_sample_action = action.sample(n=40, random_state=42)
random_sample_action.head(10)
```

```python
# Getting a random sample of 40 movies containing "adventure"
adventure = df_dropped_na[df_dropped_na['genres'].str.contains('Adventure')]
random_sample_adventure = adventure.sample(n=40, random_state=42)
random_sample_adventure.head()
```

```python
# Getting a random sample of 40 movies containing "fantasy"
fantasy = df_dropped_na[df_dropped_na['genres'].str.contains('Fantasy')]
random_sample_fantasy = fantasy.sample(n=40, random_state=42)
random_sample_fantasy.head()
```

```python
comedy = df_dropped_na[df_dropped_na['genres'].str.contains('Comedy')]
random_sample_comedy = comedy.sample(n=40, random_state=42)
random_sample_comedy.head()
```

```python
# Getting a random sample of 40 movies containing "drama"
drama = df_dropped_na[df_dropped_na['genres'].str.contains('Drama')]
random_sample_drama = drama.sample(n=40, random_state=42)
random_sample_drama.head()
```

```python
romance = df_dropped_na[df_dropped_na['genres'].str.contains('Romance')]
random_sample_romance = romance.sample(n=40, random_state=42)
random_sample_romance.head()
```

```python
# 2 - sample t test between profits of action vs romance movies
stats.ttest_ind(random_sample_action['domestic_gross'], random_sample_romance['domestic_gross'], equal_var=False)

# Print results
#print("p-value:", p_value)

# Check significance
#alpha = 0.05
#if p_value < alpha:
#    print("Reject null hypothesis: There are significant differences between the groups.")
#else:
#    print("Fail to reject null hypothesis: There are no significant differences between the groups.")
```

```python
from scipy.stats import f_oneway

# Assuming you have arrays or lists containing the data for each group
# For example, data_group1, data_group2, data_group3 represent the data for three different groups

# Perform ANOVA
f_statistic, p_value = f_oneway(random_sample_drama['domestic_gross'], random_sample_comedy['domestic_gross'], random_sample_action['domestic_gross'], random_sample_romance['domestic_gross'])

# Print results
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Check significance
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There are significant differences between the groups.")
else:
    print("Fail to reject null hypothesis: There are no significant differences between the groups.")
```

```python
#creating a dictionary with total profits for all genres with a value of 1

genre_profits = {}

for genre in genre_columns:
    genre_profit_sum = tbdf.loc[tbdf[genre] == 1, 'domestic_Profit'].sum()
    genre_profits[genre] = genre_profit_sum

print("Total profit for each genre:")
for genre, profit_sum in genre_profits.items():
    print(genre, ":", profit_sum)
```

```python
# Convert that dictionary to its own DataFrame
genre_profit_df = pd.DataFrame(list(genre_profits.items()), columns=['Genre', 'Total_Profit'])

print(genre_profit_df)
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame containing movie data
# Assuming 'domestic_Profit' is the column containing domestic profits
# Assuming 'release_year' is the column containing the release year
# Assuming 'studio' is the column containing the studio information

# Filter the DataFrame for movies produced by studio BV
bv_movies = df[df['studio'] == 'BV']

# Group the filtered DataFrame by release year and calculate the mean profit for each year
mean_profit_per_year = bv_movies.groupby('year')['domestic_Profit'].mean()

# Convert the resulting series to DataFrame if needed
mean_profit_per_year_df = mean_profit_per_year.reset_index()

# Set the mean profit per year for studio BV as a new variable
mean_profit_bv = mean_profit_per_year_df
mean_profit_bv['domestic_Profit'] = mean_profit_bv['domestic_Profit'] / 10000000

# Display the result
print(mean_profit_bv)
```

```python
# Reshape X and Y to have compatible shapes
X = np.array(mean_profit_bv['year'])  # Assuming disney_dates is a list of dates
Y = np.array(mean_profit_bv['domestic_Profit'])
X_plus_c = sm.add_constant(X)
# you have to add the y-int to the equation manually in statsmodels (put X and not Y)
#Then, call uppercase OLS and the method fit:
model = sm.OLS(Y, X_plus_c).fit()
model.summary()
```

```python
#use green as color for individual points 
plt.plot(X, Y, 'o', color='black') 
 
#obtain m (slope) and b(intercept) of linear regression line 
m, b = np.polyfit(X, Y, 1) 
 
#use red as color for regression line 
plt.plot(X, m*X+b, color='blue')

plt.xlabel('Year')
plt.ylabel('Mean Profits in Billions')
plt.title('Mean Profits of Disney Over Time');
```

```python
def sse(m):
    # sum of squared errors
    line = m*X
    err = sum(X**2 for x in [line - model.predict(X_plus_c)])
    return sum(err)
```
